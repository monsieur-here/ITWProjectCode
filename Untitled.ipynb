```python
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn import metrics

Q1 = data['horsepower'].quantile(0.25)
Q3 = data['horsepower'].quantile(0.75)
IQR = Q3 - Q1
filter = (data['horsepower'] >= Q1 - 1.5 * IQR) & (data['horsepower'] <= Q3 + 1.5 *IQR)
data=data.loc[filter]  


Q1 = data['cylinders'].quantile(0.25)
Q3 = data['cylinders'].quantile(0.75)
IQR = Q3 - Q1
filter = (data['cylinders'] >= Q1 - 1.5 * IQR) & (data['cylinders'] <= Q3 + 1.5 *IQR)
data=data.loc[filter]  


Q1 = data['displacement'].quantile(0.25)
Q3 = data['displacement'].quantile(0.75)
IQR = Q3 - Q1
filter = (data['displacement'] >= Q1 - 1.5 * IQR) & (data['displacement'] <= Q3 + 1.5 *IQR)
data=data.loc[filter]  


Q1 = data['weight'].quantile(0.25)
Q3 = data['weight'].quantile(0.75)
IQR = Q3 - Q1
filter = (data['weight'] >= Q1 - 1.5 * IQR) & (data['weight'] <= Q3 + 1.5 *IQR)
data=data.loc[filter]  


Q1 = data['acceleration'].quantile(0.25)
Q3 = data['acceleration'].quantile(0.75)
IQR = Q3 - Q1
filter = (data['acceleration'] >= Q1 - 1.5 * IQR) & (data['acceleration'] <= Q3 + 1.5 *IQR)
data=data.loc[filter]  


Q1 = data['model year'].quantile(0.25)
Q3 = data['model year'].quantile(0.75)
IQR = Q3 - Q1
filter = (data['model year'] >= Q1 - 1.5 * IQR) & (data['model year'] <= Q3 + 1.5 *IQR)
data=data.loc[filter]  


Q1 = data['origin'].quantile(0.25)
Q3 = data['origin'].quantile(0.75)
IQR = Q3 - Q1
filter = (data['origin'] >= Q1 - 1.5 * IQR) & (data['origin'] <= Q3 + 1.5 *IQR)
data=data.loc[filter]  


factors = ['displacement','horsepower','acceleration','weight','origin','model year']
X = pd.DataFrame(data[factors].copy())
y = data['mpg'].copy()

X = StandardScaler().fit_transform(X)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=932377)
# X_train.shape[0] == y_train.shape[0]
regressor = LinearRegression()
regressor.get_params()
regressor.fit(X_train,y_train)
y_predicted = regressor.predict(X_test)
rmse = sqrt(metrics.mean_squared_error(y_true=y_test,y_pred=y_predicted))  
print(rmse)
```

```python
from sklearn.metrics import r2_score
r2_score(y_true=y_test,y_pred=y_predicted)
```
